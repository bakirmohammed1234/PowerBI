```python
#5.	Ajouter une nouvelle colonne nomme "Age" qui calcule l'ge des tudiants en se basant sur leurs dates de naissance "Date_Naissance".
from datetime import date

def calculer_age(date_naissance):
    aujourd_hui = date.today()
    age = aujourd_hui.year - date_naissance.year - ((aujourd_hui.month, aujourd_hui.day) < (date_naissance.month, date_naissance.day))
    return age


date_naissance = date(2002, 6, 25)
print(date_naissance)
#print("ge :", calculer_age(date_naissance))
```

```python
def calculer_age(date_naissance):
    aujourd_hui = date.today()
    age = aujourd_hui.year - date_naissance.year - ((aujourd_hui.month, aujourd_hui.day) < (date_naissance.month, date_naissance.day))
    return age
    
```

```python
ages=[]
for i in range(len(df)):
    datenaissance = df["Date_Naissance"].iloc[i]
   
    age=calculer_age(datenaissance )
    ages.append(age)

ages
```

```python
df.insert(3,"Age",ages,True)
df
```

```python
emails=[]
pattern="@usmba.ac.ma"
for i in range(len(df)):
    prenom = df["Prenom"].iloc[i].replace(" ", "")
    nom = df["Nom"].iloc[i].replace(" ", "")
    last=prenom+"."+nom+pattern
    emails.append(last)
```

```python
emails
```

```python
df.insert(4,"Email",emails,True)
df
```

```python
accepts = []
for i in range(len(df)):
    if (df["Age"].iloc[i] <= 21) and (df["Ville_Naissance"].iloc[i] in ["FES", "MEKNES"]):
        accepts.append("Yes")
    else:
        accepts.append("No")
```

```python
accepts
```

```python
df.insert(6,"Accepted",accepts,True)
df
```
